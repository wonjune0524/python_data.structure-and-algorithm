```python
# Numpy, Pandas, Matplotlib을 이용해서 데이터 시각화
# 필요한 library
# Numpy : 수치 데이터를 조작
# Pandas : 데이터 분석
# Matplotlib : 그래프
import numpy as np    # as ~ : 별칭
import pandas as pd
import matplotlib.pyplot as plt

# 랜덤 생성기 초기화
np.random.seed(0)

data_size = 100
x = np.linspace(0, 10, data_size)   # 0부터 10까지 100개 만큼의 데이터 생성 / linspace는 균등한 간격의 숫자 생성
y = np.sin(x) + np.random.normal(scale = 0.5, size = data_size)   # 평균이 0이고, 표준편차가 0.5 / 고르게 분포된 x값에 대해 랜덤한 y값을 출력

df = pd.DataFrame({
    'X' : x,
    'Y' : y
})
print(df)
```

```python
plt.style.use('seaborn')    # 그래프 스타일 설정
plt.figure(figsize = (10, 6))   # 그래프의 크기 설정
plt.plot(df['X'], df['Y'], label = "Random Pos Graph", color = 'b', marker = 'o')   # 꺾은선 그래프 생성 / marker는 점의 스타일

plt.title("Random Pos Graph")   # 그래프 제목
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()    # 이는 그래프 오른쪽 위의 Random Pos Graph를 표시해줌.
plt.show()    # 설정된 그래프 출력
```

```python
# numpy image 예제 1
from PIL import Image
import numpy as np

item = Image.open(path + "item.png")
image_array = np.array(item)
print(image_array.shape)
print(image_array[350, 350])    # R, G, B, A값 출력 (색깔)
bright_image_array = np.clip(image_array[:, :, :3].astype(np.int16) + 50, 0, 255).astype(np.uint8)    # uint8 => unsigned int 8 bit
print(bright_image_array[350, 350])
print(bright_image_array.shape)   # 4에서 3이됨. 색깔 가짓수 하나 줄어듬.
bright_image_array = np.concatenate((bright_image_array, image_array[:, :, 3:]), axis = 2)
print(bright_image_array.shape)
bright_image = Image.fromarray(bright_image_array)
plt.imshow(bright_image)
```

```python
print(image_array.shape)

r_channel = image_array[:, :, 0]
g_channel = image_array[:, :, 1]
b_channel = image_array[:, :, 2]
alpha_channel = image_array[:, :, 3]

r_channel = np.clip(r_channel.astype(np.int16) + 50, 0, 255).astype(np.uint8)
g_channel = np.clip(g_channel.astype(np.int16) - 50, 0, 255).astype(np.uint8)   # g_channel값을 조절함으로서 이미지에 초록빛이 더 돌게함
b_channel = np.clip(b_channel.astype(np.int16) - 50, 0, 255).astype(np.uint8)
merged_array = np.stack([r_channel, g_channel, b_channel, alpha_channel], axis = 2)
merged_image = Image.fromarray(merged_array)
plt.imshow(merged_image)
```

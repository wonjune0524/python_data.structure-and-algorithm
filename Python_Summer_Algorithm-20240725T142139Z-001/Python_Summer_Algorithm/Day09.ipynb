```python
series = pd.Series([10, 20, 30, 40, 50], index = ['a', 'b', 'c', 'd', 'e'])
print(series)
print(series['a'])    # index 이름
print(series[0])      # positional index
```

```python
print(series['b' : 'd'])    # index명으로 slicing 가능, 이는 종료 index도 포함해서 return 해줌.

print(series[1 : 4])    # positional index는 기존이랑 같음
```

```python
series1 = pd.Series([1, 2, 3, 4])
series2 = pd.Series([10, 20, 30, 40])
print(series1 + series2)
print(series2 * 2)
```

```python
series2 = pd.Series([10, 20, 30, 40])
print(series2.mean())   # 평균
print()
print(series2.std())    # 표준편차
print()
print(series2.describe())   # 데이터의 전반적인 분포를 이해할 수 있는 기본 통계 정보
```

```python
print(series2[series2 > 20])
```

```python
# 데이터 조작 및 정제
series = pd.Series([10, 20, 30, 40, 50], index = ['a', 'b', 'c', 'd', 'e'])
# 데이터 변경 및 추가
series['a'] = 100
series['f'] = 60
print(series)
```

```python
# 학생 성적 데이터
scores = pd.Series([85, 92, 78, 90, 95], index = ['John', 'Emma', 'Michael', 'Sophia', 'James'])

# 평균 출력
print(scores.mean())
print()


# 조건 필터링 : 성적 90점 이상
print(scores[scores >= 90])
```

```python
df_m = pd.read_excel(path + 'pop2024.xlsx', \
                     skiprows = 3, index_col = '행정기관', usecols = 'B, R : AB')
df_m
```

```python
df_m.columns = df_m.columns.str.replace('.1', '')
df_m
```

```python
# 단위 구분 기호 ','로 인해 인구수가 정수가 아닌 문자 형태로 되어있기 때문에 변환해준다.
for col in df_m.columns:
  if df_m[col].dtype == 'object':
    df_m[col] = df_m[col].str.replace(',', '').astype('int', errors = 'ignore')

df_m
```

```python
df_w = pd.read_excel(path + 'pop2024.xlsx', \
                    skiprows = 3, index_col = '행정기관', usecols = 'B, AE : AO')
df_w.columns = df_w.columns.str.replace('.2', '')
for col in df_w.columns:
  if df_w[col].dtype == 'object':
    df_w[col] = df_w[col].str.replace(',', '').astype('int', errors = 'ignore')

df_w
```

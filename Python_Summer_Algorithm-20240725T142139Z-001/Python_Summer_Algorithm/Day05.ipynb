```python
# 해시 테이블 : 데이터 사이즈가 정해져 있을 때의 저장 알고리즘
# 충돌 방지 방법
# - 체이닝 : 해시 키가 겹쳐도 연이어서 저장
# - 개방 주소법 : 해시 키가 충돌하면 다른 빈 공간을 찾아 값을 저장

# 체이닝
class HashTable:
  def __init__(self, size = 10):
    self.size = size
    self.table = [[] for _ in range(size)]

  def _hash_function(self, key):
    return sum(ord(char) for char in key) % self.size

  def insert(self, key, value):
    index = self._hash_function(key)
    for k in self.table[index]:
      if k[0] == key:   # key값이 동일할 때(k[0]은 key, k[1]은 value)
        k[1] = value
        return
    self.table[index].append([key, value])

  def display(self):
    for i, bucket in enumerate(self.table):
      print(f"Index {i} : {bucket}")

hash_table = HashTable()
hash_table.insert("apple", 10)
hash_table.insert("banana", 11)
hash_table.display()
print("==")
hash_table.insert("banana", 13)
hash_table.insert("iherry", 15)
hash_table.display()
```

```python
# 개방 주소법(Open Addressing) : 충돌 발생 시 빈 공간을 찾아 이동
# 선형 탐사법(Linear Probing) : 충돌 발생 시 고정된 간격을 이용해 다른 슬롯을 탐사
class LinearProbingHashTable:
  def __init__(self, size = 10):
    self.size = size
    self.table = [None] * size

  def _hash_function(self, key):
    return sum(ord(char) for char in key) % self.size

  def insert(self, key, value):
    index = self._hash_function(key)
    original_index = index
    while self.table[index] is not None and self.table[index][0] != key:
      index = (index + 1) % self.size
      if index == original_index:
        raise Exception("Hash table is full")
    self.table[index] = (key, value)    # 튜플 : 리스트랑 방식이 비슷하지만 수정/삭제는 불가능한 자료구조

  def display(self):
    for i, kv in enumerate(self.table):
      print(f"Index {i} : {kv}")

lph = LinearProbingHashTable()
lph.insert("apple", 11)
lph.display()
print("==")
lph.insert("banana", 12)
lph.display()
print("==")
lph.insert("iherry", 13)
lph.display()
print("==")
```

```python
# 이차(제곱) 탐사법(Quadratic Probing) : 제곱된 간격을 이용해 다음 슬롯을 탐사
class QuadraticProbingHashTable:
  def __init__(self, size = 10):
    self.size = size
    self.table = [None] * size

  def _hash_function(self, key):
    return sum(ord(char) for char in key) % self.size

  def insert(self, key, value):
    index = self._hash_function(key)
    original_index = index
    i = 1
    while self.table[index] is not None and self.table[index][0] != key:
      index = (original_index + i * i) % self.size
      i += 1
      if i == self.size:
        raise Exception("Hash table is full")
    self.table[index] = (key, value)

  def display(self):
    for i, kv in enumerate(self.table):
      print(f"Index {i} : {kv}")

qph = QuadraticProbingHashTable()
qph.insert("apple", 10)
qph.display()
print("==")
qph.insert("banana", 11)
qph.display()
print("==")
qph.insert("iherry", 12)
qph.display()
print("==")

# 선형 탐사보다 이차 탐사가 더 나은 이유
# - 보다 균일한 분포 : 충돌 가능성 감소
```

```python
# Tree : 계층적인 구조를 가진 자료구조, 루트 노드에서 시작해 하위 노드로 연결됨
"""
  루트 노드 : 트리의 최상위 노드
  부모 노드 : 특정 노드의 상위 노드
  자식 노드 : 특정 노드의 하위 노드
  형제 노드 : 같은 부모 노드를 공유하는 노드들
  단말 노드(Leaf Node) : 자식 노드가 없는 노드
  차수 : 노드가 가진 자식 노드의 수
  레벨 : 루트에서부터의 거리(루트 노드 레벨 = 0)
"""
```

```python
# 이진트리 : 각 노드가 최대 두 개의 자식 노드를 가지는 트리
# 이진탐색트리(Binary Search Tree):
#        왼쪽 서브트리 모든 값이 루트 값

# 트리노드 구현 : 트리의 각 노드를 나타내는 클래스
class TreeNode:
  def __init__(self, key):
    self.key = key
    self.left = None    # 왼쪽 자식 노드
    self.right = None   # 오른쪽 자식 노드

class BinaryTree:   # 이진 탐색 트리를 관리하는 클래스
  def __init__(self):
    self.root = None    # 루트 노드

# 트리 노드 삽입 로직 : 새로운 값을 삽입하여 트리를 구축하는 함수
  def insert(self, key):
    if not self.root:   # None 일 때(트리가 비어있을 때)
      self.root = TreeNode(key)   # root는 treenode에 관한 변수 / 새로운 노드를 루트로 설정
    else:
      self.insert_recursive(self.root, key) # 트리에 재귀적으로 값을 삽입

  def insert_recursive(self, node, key):    # 새로 추가할 노드의 부모 노드, 새로 추가할 값
    if key < node.key:                      # 새로 추가할 자식 노드 값이 부모 노드보다 작으면
      if node.left is None:                 # 왼쪽 자식이 비어있으면
        node.left = TreeNode(key)           # 새 노드를 왼쪽에 삽입
      else:
        self.insert_recursive(node.left, key)   # 왼쪽 서브트리로 재귀
    else:                                   # 자식노드 값이 부모노드보다 크거나 같으면
      if node.right is None:                # 오른쪽 자식이 비어있으면
        node.right = TreeNode(key)          # 새 노드를 오른쪽에 삽입
      else:
        self.insert_recursive(node.right, key)  # 오른쪽 서브트리로 재귀

# 트리 출력(display) 기능 : 트리의 구조를 시각적으로 출력하는 함수
  def display(self):                    # 트리의 현재 상태를 시각적으로 출력하는 함수
    lines, *_ = self.display_recursive(self.root)
    for line in lines:
      print(line)

  def display_recursive(self, node):    # 트리를 문자열로 변환한 후, 각 줄을 출력하는 함수
    if node is None:
      return [], 0, 0     # 빈 리스트, 너비 = 0, 높이 = 0 반환 / 재귀의 종료 조건
    # left_lines와 right_lines는 각각 왼쪽과 오른쪽 서브트리의 문자열 리스트임.
    # 각 서브트리의 너비와 높이를 구하고, 공백을 채워서 트리의 형태를 맞추고 정렬함.
    left_lines, left_width, left_height = self.display_recursive(node.left)
    right_lines, right_width, right_height = self.display_recursive(node.right)
    node_key_width = len(str(node.key))   # 정수를 문자열로 변환하고 길이 계산
    total_width = left_width + node_key_width + right_width
    total_height = max(left_height, right_height) + 1
    middle = node_key_width // 2    # 트리의 각 노드가 가운데 정렬되도록 하기 위한 현재 노드의 중간 위치 계산

    # 서브트리의 높이가 다를 수 있기 때문에 전체 높이에 낮은 서브트리의 높이를 맞춰줌
    while len(left_lines) < total_height:
      left_lines.append(" " * left_width)
    while len(right_lines) < total_height:
      right_lines.append(" " * right_width)

    middle_line = " " * left_width + str(node.key) + " " * right_width    # 현재 노드를 중앙에 배치
    lines = [middle_line]
    for i in range(total_height):
      lines.append(left_lines[i] + " " * node_key_width + right_lines[i])
    return lines, total_width, total_height

binary_tree = BinaryTree()
keys = [10, 5, 15, 3, 7, 12, 18]
for key in keys:
  binary_tree.insert(key)

# 왼쪽 서브트리(3, 5, 7)와 오른쪽 서브트리(12, 15, 18)로 나뉘어서 출력됨.
# 5와 15는 각 서브트리의 middle값이다.
binary_tree.display()
```

```python
# 이진트리 순회
# 전위 순회(Preorder Traversal) : 부모 노드 -> 왼쪽 자식 노드 -> 오른쪽 자식 노드
# 중위 순회(Inorder Traversal) : 왼쪽 자식 노드 -> 부모 노드 -> 오른쪽 자식 노드
# 후위 순회(Postorder Traversal) : 왼쪽 자식 노드 -> 오른쪽 자식 노드 -> 부모 노드
class BinaryTreeTraversal(BinaryTree):
  def __init__(self):
    super().__init__()    # BinaryTree 클래스를 상속받고 있음

  def preorder_traversal(self, root):   # 전위 순회
    if root:
      print(root.key, end = " ")        # 부모 노드
      self.preorder_traversal(root.left)  # 왼쪽 노드
      self.preorder_traversal(root.right) # 오른쪽 노드

  def inorder_traversal(self, root):    # 중위 순회
    if root:
      self.inorder_traversal(root.left)   # 왼쪽 노드
      print(root.key, end = " ")          # 부모 노드
      self.inorder_traversal(root.right)  # 오른쪽 노드

  def postorder_traversal(self, root):  # 후위 순회
    if root:
      self.postorder_traversal(root.left)  # 왼쪽 노드
      self.postorder_traversal(root.right) # 오른쪽 노드
      print(root.key, end = " ")          # 부모 노드

tree_to_traversal = BinaryTreeTraversal()
keys = [10, 5, 15, 3, 7, 12, 18]
for k in keys:
  tree_to_traversal.insert(k)

# 이진 트리 형태로 값들을 정렬한 후, 순회한다.
# 위의 이진 탐색 트리의 출력값을 참고하여 이 코드의 출력값을 이해하자.
root = tree_to_traversal.root
tree_to_traversal.preorder_traversal(root)
print()
tree_to_traversal.inorder_traversal(root)
print()
tree_to_traversal.postorder_traversal(root)
print()
```

```python
class BinarySearchTree(BinaryTree):   # 이진탐색트리
  def __init__(self):
    super().__init__()    # BinaryTree 클래스를 상속받고 있음

  def search(self, key):
    return self.search_recursive(self.root, key)

  def search_recursive(self, node, key):
    if node is None or node.key == key:
      return node
    elif key < node.key:
      return self.search_recursive(node.left, key)
    elif key > node.key:
      return self.search_recursive(node.right, key)

bst = BinarySearchTree()
bst.insert(30)
bst.insert(10)
bst.insert(20)
bst.insert(5)
bst.insert(15)
bst.display()

node = bst.search(20)
print(node.key)
print(node.left.key)
```

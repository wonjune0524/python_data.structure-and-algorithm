```python
def permute(s, answer):
  if len(s) == 0:
    print(answer)
    return
  for i in range(len(s)):
    ch = s[i]
    left_substr = s[0:i]
    right_substr = s[i + 1:]
    rest = left_substr + right_substr
    permute(rest, answer + ch)

text = "ABC"
permute(text, "")
```

```python
# 큐 : 선입선출의 방식을 가지는 자료구조 (출,입구가 따로 있음)
class Queue:
  def __init__(self):
    self.queue = []

  def is_empty(self):
    return len(self.queue) == 0

  def enqueue(self, item):    # 삽입 메서드 / 큐에 넣는 행동
    self.queue.append(item)
    print(f"Enqueued : {item}")

  def dequeue(self):          # 삭제 메서드 / 큐에서 빼는 행동
    if self.is_empty():
      print("Queue is empty!")
      return None
    item = self.queue.pop(0)    # pop 메서드의 괄호 안은 요소의 인덱스 값이 온다. 괄호가 비어있을 경우, 맨 뒤의 값을 pop한다.
    print(f"Dequeued : {item}")
    return item   # 반환할 필요 없으면 경우에 따라 쓰지 않아도 된다.

  def print(self):
    print(f"Queue : {self.queue}")

q = Queue()
q.enqueue(1)
q.print()
q.enqueue(2)
q.print()
q.enqueue(3)
q.print()
q.dequeue()   # 요소 하나씩 제거
q.print()
q.dequeue()
q.print()
```

```python
# 데크(dequeue) : 양쪽 끝에서 삽입과 삭제가 가능한 자료구조
# front : 앞, rear : 뒤
class Deque:
  def __init__(self):
    self.deque = []

  def is_empty(self):
    return len(self.deque) == 0

  def add_front(self, item):
    self.deque.insert(0, item)    # insert 메서드는 리스트에서 지정한 index에 item을 넣게 해줌.
    print(f"Added to front : {item}")

  def remove_front(self):
    if self.is_empty():
      print("Deque is empty!")
      return None
    item = self.deque.pop(0)
    print(f"Removed from front : {item}")
    return item

  def add_rear(self, item):
    self.deque.append(item)
    print(f"Added to rear : {item}")

  def remove_rear(self):
    if self.is_empty():
      print("Deque is empty!")
      return None
    item = self.deque.pop()    # pop 함수는 매개변수가 없으면 무조건 뒤에서 요소를 제거
    print(f"Removed from rear : {item}")
    return item

  def print(self):
    print(f"Current Deque : {self.deque}")

d = Deque()
d.add_front(1)    # 앞에 1
d.print()
d.add_rear(2)     # 뒤에 2
d.print()
d.add_front(3)    # 앞에 3
d.print()
d.add_rear(4)     # 뒤에 4
d.print()
d.remove_front()
d.print()
d.remove_rear()
d.print()
```

```python
# 원형 큐
class CircularQueue:
  def __init__(self, size):
    self.size = size    # 리스트 사이즈를 의미함.
    self.queue = [None] * size    # [None, None, None, None, None] => 이런 식으로 queue가 생김.
    self.front = self.rear = -1   # 각각 앞과 뒤의 index를 가리킴.

  def is_full(self):
    # rear + 1을 size로 나눈 나머지가 front랑 같을 때 : 큐가 가득 찬 상태 / 이는 front와 rear이 겹치는 상태.
    return (self.rear + 1) % self.size == self.front    # 원형 모양의 큐에서 시작점이 front, 끝점이 rear이라 생각해보자.
                                                        # ex) front = 1, size = 8, rear = 8이라면, rear + 1이 9가 되며 front 위에 겹쳐지게 된다.
                                                        # 이를 생각해봤을 때, 해당 계산식이 True로 나오게 된다.

  def is_empty(self):
    return self.front == -1

  def enqueue(self, item):
    if self.is_full():
      print("Queue is full!")
      return
    if self.is_empty():
      self.front = 0
    self.rear = (self.rear + 1) % self.size
    self.queue[self.rear] = item
    print(f"Enqueued : {item}")

  def dequeue(self):      # 원형 큐에서 dequeue는 물리적으로 요소를 없애는 것이 아닌, 단지 front 포인터를 이동시키는 방식으로 작동한다. / 원형 큐는 pop 메서드 사용불가
    if self.is_empty():
      print("Queue is empty!!")
      return None   # dequeue 될 때 요소가 튀어나오면서 return 되기 때문에 None으로 설정해줌.
    item = self.queue[self.front]
    if self.front == self.rear:   # 비어있는 것으로 인식 / front는 요소를 지우기 위해, rear는 요소를 추가하기 위해 사용되고 있다.
      self.front = self.rear = -1
      # a = b = c = 5     # a, b, c에 동시에 5를 대입. / 과 같은 논리이다.
    else:
      # self.queue[self.front] = None   => 불필요한 연산 줄이기
      self.front = (self.front + 1) % self.size
    print(f"Dequeued : {item}")
    return item

  def print(self):
    if self.is_empty():
      print("Queue is empty!")
      return
    if self.rear >= self.front:
      print("Queue elements : ", end ="")
      for i in range(self.front, self.rear + 1):
        print(self.queue[i], end =" ")
      print()   # 줄바꿈
    else:
      print("Queue elements : ", end ="")
      for i in range(self.front, self.size):
        print(self.queue[i], end =" ")
      for i in range(0, self.rear + 1):
        print(self.queue[i], end =" ")
      print()

cq = CircularQueue(5)
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)
cq.enqueue(4)
cq.enqueue(5)
cq.dequeue()
cq.dequeue()
cq.enqueue(6)
cq.enqueue(7)
cq.print()
```

# 일반 큐와 원형 큐의 차이점:
- 일반 큐는 요소가 제거되면서 생긴 빈 공간을 뒤의 요소들이 앞으로 당김으로써 메꿔야 한다.
이와 달리 원형 큐는 그러지 않아도 된다.
+ 원형 큐는 pop 메서드를 사용할 수 없다.

```python
# 괄호의 균형 검삭
def check_parenthesis_balance(expression):
  stack = []
  for char in expression:
    if char in "({[":
      stack.append(char)    # stack에 push
    elif char in ")}]":
      if not stack:   # stack이 비어있는지 확인 / 비어있을 경우, not stack => True
        return False    # 따라서 여기서 False 반환함.
      top = stack.pop()   # stack에서 제거하며 return
      if char == ")" and top != "(":
        return False
      elif char == "}" and top != "{":
        return False
      elif char == "]" and top != "[":
        return False
  return not stack
# ({[]})    출력 : True => 짝이 맞음
# ([]       출력 : False => 짝이 맞지 않음
print(check_parenthesis_balance("({[]})"))
print(check_parenthesis_balance("([]"))
```

```python
def reverse_string(s):
  stack = []
  for char in s:
    stack.append(char)
  reversed_str = ""
  while stack:    # 비어있지 않으면 True
    reversed_str += stack.pop()
  return reversed_str

print(reverse_string("hello"))
```

```python
# 큐 : 선입선출의 방식을 가지는 자료구조 (출,입구가 따로 있음)
class Queue:
  def __init__(self):
    self.queue = []

  def is_empty(self):
    return len(self.queue) == 0

  def enqueue(self, item):    # 삽입 메서드 / 큐에 넣는 행동
    self.queue.append(item)
    print(f"Enqueued : {item}")

  def dequeue(self):          # 삭제 메서드 / 큐에서 빼는 행동
    if self.is_empty():
      print("Queue is empty!")
      return None
    item = self.queue.pop(0)
    print(f"Dequeued : {item}")
    return item   # 반환할 필요 없으면 경우에 따라 쓰지 않아도 된다.

  def print(self):
    print(f"Queue : {self.queue}")

q = Queue()

# 큐를 이용해서 이진수 생성
def gen_bin_nums(n):
  q.enqueue("1")
  r = Queue()
  while len(r.queue) < n:
    current = q.dequeue()
    r.enqueue(current)
    q.enqueue(current + "0")
    q.enqueue(current + "1")
    print("")
  return r

print(gen_bin_nums(5).print())
q.print()
```

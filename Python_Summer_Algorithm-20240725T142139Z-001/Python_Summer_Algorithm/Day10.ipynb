```python
# 막대 그래프
labels = ['Amy', 'John', 'Emma']
values = [155, 177, 166]

plt.bar(labels, values, color = 'r', alpha= 0.6)
# (세로) 막대 그래프 : x축에 label, y축 value들로 막대 그래프를 구성
# color = 색상, alpha = 투명도(0 ~ 1)
```

```python
colors = ['r', 'g', 'b']
plt.bar(labels, values, color = colors)
plt.ylim(150, 180)    # y축 숫자 최저선, 최고선
```

```python
plt.bar(labels, values, width = 0.3)
newlabel = ['A', 'B', 'C']
plt.xticks(labels, newlabel, rotation = 45)
plt.yticks(rotation = 45)
```

```python
# 세로 막대 그래프
plt.barh(labels, values)
plt.xlim(150, 180)
```

```python
# 각 막대 그래프를 list로 return
bar = plt.barh(labels, values)
print(bar)
bar[2].set_hatch('.')
bar[1].set_hatch('x')
bar[0].set_hatch('/')
# / \\ | - + o *
```

```python
# 원 그래프
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C++', 'ETC']
explode = [0.2, 0.1, 0.3, 0, 0, 0]    # 각 섹션이 중심에서 얼마나 떨어질지

plt.pie(values, labels = labels, autopct = '%.2f%%', startangle = 90, counterclock = False, explode = explode)
# autopct : 소수점 둘째 자리 실수로 표현, 끝에 %를 붙이는 방식으로 값을 표현하겠다.
# startangle : 원 그래프의 시작 각도, 0도는 제 1사분면 x축
# counterclock : 기본값이 반시계
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C++', 'ETC']
explode = [0.05] * 6
colors = ['#030BA6', '#2384D9', '#03A696', '#D94436', '#8C0F0F', '#FFFFFF']
wedgeprops = {'width' : 0.8, 'edgecolor' : 'gray', 'linewidth' : 2}

plt.pie(values, labels = labels, explode = explode, colors = colors, wedgeprops = wedgeprops)
plt.legend(loc = (1.2, 0.3), title = '언어 선호도')
plt.show()
```

```python
gender = df.groupby('Sex')['Survived'].value_counts(normalize = True)
gender
```

```python
gender = df.groupby('Sex')['Survived'].value_counts(normalize = True).unstack()
# unstack : 피벗 테이블로 생성
gender
```

```python
gender.plot(kind = 'bar', stacked = True, color = ['lightcoral', 'skyblue'])
# stacked = True : 누적 막대 그래프
plt.title('성별에 따른 생존율')
plt.ylabel('생존율')
plt.xlabel('성별')
plt.xticks(rotation = 0)
plt.legend(['Not Survived', 'Survived'], title = '상태')
plt.show()
```

```python
plt.hist(df['Age'], bins = 30, alpha = 0.7, color = 'blue')
# bins = 전체 나이 구간을 30개 구간으로 나누겠다.
plt.title('승객의 나이 분포')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 상자 그림(Box Plot) 그래프
# 데이터의 분포를 시각적으로 나타내는 그래프
import seaborn as sns   # 데이터 시각화를 위한 라이브러리, matplotlib를 기반으로 좀 더 추가된 디테일의 그래프를 그릴 수 있게 해줌

sns.boxplot(x = 'Embarked', y = 'Fare', data = df, palette = 'Set3')
plt.title('승선지에 따른 요금 분포')
plt.xlabel('승선지')
plt.ylabel('요금')
plt.show()
```

```python
sns.scatterplot(x = 'Age', y = 'Fare', hue = 'Embarked', data = df, palette = 'Set3')
plt.title('승선지에 따른 나이와 요금 분포')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

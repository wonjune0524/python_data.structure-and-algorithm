```python
# Singly Linked list / 여기에서 리스트는 앞의 리스트와 조금 다른 개념
# 리스트나 튜플같은 자료구조는 딱 정해진 배열구조를 갖는다. / 실제로 요소를 추가할 때, 밀어냄으로써 새로운 요소를 채우게 된다.
# 이는 메모리상 전통적 방식임. / 자리가 일렬로 정해져있음 ㅁㅁㅁㅁ 이렇게
# 링크드 리스트는 Node 라고 하는 단위체를 기준으로 한칸씩 연결된 기차모양을 띔. ㅁ-ㅁ-ㅁ-ㅁ-ㅁ 자리와 다음 주소를 가리키는 변수가 따로 있다,

class Node:
  def __init__(self, data):
    self.data = data
    self.next = None    # next 변수는 다음 노드에 대한 참조 / 다음 노드를 가리키는 포인터임.

class SinglyLinkedList:
  def __init__(self):
    self.head = None    # head 인스턴스 변수 : head = 리스트의 첫 번째 노드를 가리킴.
                        # None = 아무것도 안 가르키는 상태
  def append(self, data):   # 맨 뒤에 요소를 추가 하는 작업
    new_node = Node(data)   # Node 객체 생성 / append 메서드를 호출할 때마다 Node 객체가 새로 생성되는 것임.
    if not self.head:     # 맨 처음에 새로 생성 될 경우 => self.head가 비어 있으면 None, 따라서 False가 출력됨. not None => True 출력.
                          # head는 메모리를 주소값으로 갖는다.
      self.head = new_node    # head가 가리키는 node가 없을 경우, new_node를 가리킨다.
      return
    last_node = self.head
    while last_node.next:     # 처음엔 last_node.next가 가리키는 node가 없기 때문에 실행되지 않고 (조금 더 정확히 말하자면 "마지막 노드를 찾을 때까지 반복, 근데 처음 이 반복문이 실행될 때에는 반복이 진행될 수 없다.")
      last_node = last_node.next  # last_node를 next를 통해 다음 node로 이동시켜줌.
    last_node.next = new_node     # 이 코드가 실행된다. last_node.next 변수에 값 할당해줌.
  def print_list(self):
    current_node = self.head
    while current_node:   # head가 None이 아니면 무조건 실행됨. / 리스트의 처음부터 끝까지 Node의 데이터를 출력함.
      print(current_node.data, end=" -> ")
      current_node = current_node.next    # current_node를 다음 node로 이동시켜줌.
    print("None")
  def prepend(self, data):  # 맨 앞에 요소를 추가하는 함수
    new_node = Node(data)   # Node 객체 생성
    new_node.next = self.head
    self.head = new_node
  def insert(self, index, data):
    if index == 0:
      self.prepend(data)
      return
    new_node = Node(data)   # Node 객체 생성
    current_node = self.head
    current_position = 0

    while current_node and current_position < index - 1:    # current_node가 True이고, 포지션이 해당 조건에 만족하면 실행
      current_node = current_node.next
      current_position += 1

    if current_node is None:
      print("Not found")
      return
    new_node.next = current_node.next   # new_node의 next가 current_node의 next를 가리키도록 함
    current_node.next = new_node        # current_node의 next가 new_node를 가리키도록 함
  def delete(self,key):
    if not self.head:   # head가 비어있을 경우, 리스트가 비어있음을 출력
      print("List is empty")
      return

    if self.head.data == key:   # 리스트의 첫 번째 Node가 삭제할 값일 때
      #temp = self.head
      self.head = self.head.next    # head가 다음 node를 가리키도록 함
      #del temp
      return

    current_node = self.head
    previous_node = None
    while current_node and current_node.data != key:
      previous_node = current_node
      current_node = current_node.next

    if current_node is None:
      print(f"Node with data {key} not found")
      return

    previous_node.next = current_node.next
    del current_node
sll = SinglyLinkedList()    # new_node로 객체가 생성됨.
sll.append(1)   # 첫 번째 Node 생성 / data 값은 1이다.
                # sll 을 만들 때 head = none이다.
                # append 함수가 실행되면 new_node 라는 변수를 얻고 그 속에 data = 1, last_node.next = None 이 들어있다.
                # 이때 sll의 head는 new_node 일 것이다. 즉, head는 이 new_node라는 인스턴스 그 자체를 가리킨다.
print(sll.head)     # Node 객체의 메모리 주소 출력
print(sll.head.data)  # 데이터로 출력되어 1이 출력
sll.append(2)         # append2를 보면 new_node를 하나 더 만들고 data = 2, last_node.next = 데이터 2의 값을 가진 new_node 이다. head가 last_node를 가리키는데, 이는 데이터가 1인 이전의 new_node를 가르키는 것이다.
print(sll.head.next.data)
sll.append(3)
sll.append(4)
sll.print_list()
sll.prepend(0)
sll.print_list()
sll.insert(4, 3.5)
sll.print_list()
sll.delete(0)
sll.delete(3.5)
sll.print_list()
```

```python
# Doubly Linked List
class Node:                     # 그림으로 이해하는 편이 훨씬 쉬움. / 스택은 선입후출 (출,입구가 같음)
  def __init__(self, data):
    self.data = data
    self.prev = None
    self.next = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None

  def append(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      return

    last_node = self.head
    while last_node.next:
      last_node = last_node.next
    last_node.next = new_node
    new_node.prev = last_node

  def print_list(self):
    current_node = self.head
    while current_node:
      print(current_node.data, end= " <-> ")
      current_node = current_node.next
    print("None")

  def insert(self, index, data):
    new_node = Node(data)
    if index == 0:
      if not self.head:
        self.head = new_node
        return
      self.head.prev = new_node
      new_node.next = self.head
      self.head = new_node    # 맨 앞에 집어넣으므로 new_node가 head가 된다.
      return

    current_node = self.head
    current_position = 0

    while current_node and current_position < index - 1:
      current_node = current_node.next
      current_position += 1

    if current_position != index - 1 or not current_node:   # node 수에 비해 index 값이 클 때 발생
      print(f"index {index} out of bounds")
      return

    new_node.next = current_node.next
    if current_node.next:
      current_node.next.prev = new_node
    current_node.next = new_node
    new_node.prev = current_node

  def delete(self, key):
    if not self.head:
      print("List is empty")
      return
    current_node = self.head
    while current_node and current_node.data != key:
      current_node = current_node.next
    if current_node is None:
      print(f"Node with data {key} not found")
      return
    if current_node.prev:
      current_node.prev.next = current_node.next
    if current_node.next:
      current_node.next.prev = current_node.prev
    if current_node == self.head:
      self.head = current_node.next

dll = DoublyLinkedList()
dll.append(2)
dll.append(4)
dll.append(5)
dll.print_list()
dll.insert(0, 1)
dll.insert(2, 3)
dll.print_list()
dll.delete(1)
dll.delete(3)
dll.print_list()
```

```python
# 재귀함수 : 함수 내에서 함수 자기 자신을 다시 호출하는 형태
# 팩토리얼
def factorial(n):
  if n == 0:
    return 1
  return n * factorial(n - 1)

print(factorial(5))
# n = 5 => return 값 = 5 * factiroal(4) => return 값 = 5 * 4 * factiroal(3)
```

```python
# 피보나치 수열
# 0 1 1 2 3 5 8 13 21
def fibonacci(n):
  if n <= 1:
    return n
  return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(7))
```

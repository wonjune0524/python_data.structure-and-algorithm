```python
# python 내장 자료구조 연습
# 딕셔너리 > 리스트
# key, value
students = [
  {"name": "Alice", "scores" : (85, 90, 88)},
  {"name": "Bob", "scores" : (78, 81, 85)},
  {"name": "Charlie", "scores" : (92, 89, 95)}
]

print(students[0])
print(students[0]["name"])
print(students[0]["scores"])
students[0]["name"] = 'Anne'
print(students)
```

```python
# 4 : 숫자 맞추기 게임
import random   # 임의의 수를 생성하는 모듈

number_to_guess = random.randint(1, 100)    # randrange 함수는 1 ~ 99까지.
# 1부터 100까지 임의의 정수를 return
attempts = 0    # 몇 번 시도했는지 count할 변수

while True:
  guess = int(input("숫자를 맞춰보세요  (1 - 100) : "))
  attempts += 1   # attempts = attempts + 1

  if guess < number_to_guess:
    print("더 큰 숫자입니다.")
  elif guess > number_to_guess:
    print("더 작은 숫자입니다.")
  else:
    print(f"축하합니다! {attempts}번 만에 맞췄습니다.")
    break
```

```python
# class 예제 : 연산자 오버로딩 (기존에 있는 연산자의 쓰는 방식을 재정의)
class Complex:
  def __init__(self, real, imag):   # 실수, 허수 / 실수 + 허수 => 복소수
    self.real = real                # 오버라이딩 => 상속받은 메서드 내용을 재정의
    self.imag = imag
  def __add__(self, other):   # instance 단위로 + 연산을 쓰게끔 해주는 연산자 오버로딩
    return Complex(self.real + other.real, self.imag + other.imag)    # 두 개의 Complex 객체를 더함.
                                                                      # 여기에서 이 더하기 메서드는 재귀함수처럼 타고 들어가는 형태로 정의되었다.
  def __str__(self):
    return f"{self.real} + {self.imag}i"

c1 = Complex(2, 3)
c2 = Complex(4, 5)
c3 = c1 + c2
print(c1 + c2)    # 이는 add 메서드와 str 메서드를 호출함. / c1.__add__(c2) , (c1 + c2).__str__()
print(c3)         # c3.__str__()
```

```python
# 다중 상속 예제
class Base1:
  def __init__(self):
    self.str1 = "Base1"
  def show_str1(self):
    return self.str1

class Base2:
  def __init__(self):
    self.str2 = "Base2"
  def show_str2(self):
    return self.str2

class Derived(Base1, Base2):
  def __init__(self):
    Base1.__init__(self)
    Base2.__init__(self)    # super().__init__(self) 이 일반적이나 지금은 다중 상속이기 때문.
  def show_all(self):
    return f"{self.show_str1()} and {self.show_str2}"

obj = Derived()
print(obj.show_all())   # Base1 and Base2 가 출력되어야 하는데 오류가 좀 있네
```

```python
# 스택 구현 / 스택은 입(출)구가 하나뿐(선입후출의 논리) , 어떤 컵에 1, 2번공이 순서대로 들어가면, 2번공을 먼저 꺼낼 수밖에 없는.
class StackOverflowError(Exception):    # Exception 클래스를 상속받음.
  pass    # 아무 작업도 수행하지 않는 키워드

class StackUnderflowError(Exception):
  pass

class Stack:
  def __init__(self, limit = 10):
    self.stack = []   # stack은 지금 비어있는 list 형태로 선언되었으므로, 괄호 안에 선언되지 않아도 됨.
    self.limit = limit
  def is_empty(self):
    return len(self.stack) == 0   # True 또는 False가 return 됨.
  def is_full(self):
    return len(self.stack) == self.limit

  def push(self, item):   # push 메서드는 특성상 item 요소를 받지만, pop 메서드는 필요없다.
    if self.is_full():
      raise StackOverflowError("Stack overflow : cannot push to full stack")
    self.stack.append(item)
  def pop(self):    # stack의 맨 위에 있는 요소를 return, 그리고 난 후 그 요소를 삭제
    if self.is_empty():
      raise StackUnderflowError("Stack underflow : cannot pop from empty stack")
    item = self.stack.pop()
    return item
  def __str__(self):
    return f"Stack : {self.stack}"

try:
  s = Stack(limit = 5)
  s.push(1)
  s.push(3)
  s.push(5)
  s.push(7)
  s.push(9)
  # s.push(11)  # overflow error
  print(s)    # s.__str__()
  s.pop()
  s.pop()
  s.pop()
  s.pop()
  s.pop()
  # s.pop()     # underflow error
  print(s)    # s.__str__()
except StackOverflowError as e:
  print(e)
except StackUnderflowError as e:
  print(e)
```

```python
class PostfixCalculator:    # 후위 계산식(연산자가 뒤에 있는 계산식)
  def __init__(self):
    self.stack =[]
  def is_number(self, token):   # 정수 확인
    try:
      int(token)
      return True
    except ValueError:    # ValueError는 직접 정의할 필요 없이 사용할 수 있는 메서드임. / 예외처리
                          # 이는 문자열을 정수로 변환할 수 없을 때, 발생하는 예외이다.
                          # 이는 내장 함수에서 인수의 유형은 올바르지만 값이 적절하지 않은 경우에 발생한다.
      return False
  def calculate(self, expr):
    tokens = expr.split()   # 공백을 기준으로 문자열을 나눔 / tokens는 문자열 리스트가 됨.

    for token in tokens:
      print("current stack : ", self.stack)
      print("token : ", token)

      if self.is_number(token) :    # token에는 정수와 연산자가 모두 있다. 이를 분류해주기 위한 조건문.
        self.stack.append(int(token))   # stack에는 정수가 들어감.
      else:
        operator = token    # operator에는 연산자가 들어감.
        operand2 = self.stack.pop()   # 정수 2
        operand1 = self.stack.pop()   # 정수 1
        result = self.operation(operator, operand1, operand2)   # operation 메서드가 아래에 정의되어 있음.
        self.stack.append(result)
    return self.stack.pop()   # for 반복문을 통해 expr의 모든 계산을 마치고 최종적으로 나온 값을 return 해준다.
  def operation(self, operator, operand1, operand2):
    if operator == '+' :
      return operand1 + operand2
    elif operator == '-':
      return operand1 - operand2
    elif operator == '*':
      return operand1 * operand2
    elif operator == '/':
      return operand1 / operand2
    else:
      raise ValueError("Unexpected token!")

expr = "8 2 / 3 - 3 2 * +"
calc = PostfixCalculator()
calc.calculate(expr)
```

```python
def hello_world():
  print("hello world")

hello_world()
hello_world()
hello_world()
# 입력값이 없는 메서드는 당연히 입력값의 영향을 받지 않음
# 입력값의 영향을 받지 않는 메서드는 시간복잡도를 상수라고 칭한다.
# 상수 시간 복잡도 = O(1)
```

```python
li = ['시간', '복잡도', '연습', 'A', 'B']
def print_each(li):
  for i in li:
    print(i)

print_each(li)
# 입력값의 영향을 받아서, 입력값 숫자대로 for문이 돌아감
# 입력값 개수를 n으로 표현해서 O(n)의 시간복잡도를 가진다 = 선형 시간 복잡도
```

```python
def print_each2(li):
  for i in li:
    for j in li:
      print(f"{j}")   # print(j)로 출력해도 똑같음.

print_each2(li)

# 시간 복잡도 표기법 : 빅 오 표기법
```
